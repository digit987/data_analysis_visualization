# 1. Age Distribution

```python
plt.figure(figsize=(12, 6))  # Set up the figure size
sns.histplot(df['Age'], bins=40, kde=True)  # Plot the histogram with KDE
plt.title('Age Distribution')  # Add a title to the plot
plt.show()  # Display the plot
```

```python
plt.figure(figsize=(12, 6))  # Define the figure size
sns.histplot(df['Income'], bins=20, kde=True)  # Plot a histogram of 'Income' with a KDE overlay
plt.title('Income Distribution')  # Set the plot title
plt.show()  # Display the plot
```

# 8. Total Time Engaged on Social Media

```python
plt.figure(figsize=(12, 6))  # Set the figure size
sns.histplot(df['Total Time Spent'], bins=40, kde=True)  # Plot a histogram with a KDE curve for 'Total Time Spent'
plt.title('Total Time Spent on Social Media')  # Add a title to the plot
plt.show()  # Display the plot
```

# 9. Impact of Time Spent on Social Media on Productivity

```python
plt.figure(figsize=(12, 6))  # Set the dimensions of the figure
sns.scatterplot(data=df, x='Total Time Spent', y='ProductivityLoss')  # Create a scatter plot of 'Total Time Spent' versus 'Productivity Loss'
plt.title('Total Time Spent vs. Productivity Loss')  # Add a title to the plot
plt.show()  # Render the plot
```

# 10. Platform-Specific Satisfaction Levels

```python
plt.figure(figsize=(12, 6))  # Define the figure size
sns.boxplot(data=df, x='Platform', y='Satisfaction')  # Create a box plot showing 'Satisfaction' levels for each 'Platform'
plt.title('Satisfaction Levels by Platform')  # Set the plot title
plt.show()  # Display the plot
```

# 11. Identifying Peak Productivity Losses by Time of Day

```python
# Calculate the average productivity loss for each frequency
frequency_productivity_loss = df.groupby('Frequency')['ProductivityLoss'].mean().reset_index()

# Sort the results to identify the highest average productivity loss
frequency_productivity_loss = frequency_productivity_loss.sort_values(by='ProductivityLoss', ascending=False)

# Print the sorted results
print(frequency_productivity_loss)

# Create a bar plot to visualize average productivity loss by frequency
plt.figure(figsize=(12, 6))  # Define the size of the figure
plt.bar(frequency_productivity_loss['Frequency'], frequency_productivity_loss['ProductivityLoss'], color=['red', 'orange', 'yellow', 'green'])  # Plot the data
plt.xlabel('Frequency')  # Label for the x-axis
plt.ylabel('Average Productivity Loss')  # Label for the y-axis
plt.title('Average Productivity Loss by Frequency')  # Title of the plot
plt.show()  # Display the plot
```

# Conclusion

**1. Total Time Spent:**

Social media usage varies widely among users, with significant differences based on age and country.
Younger users, especially those aged 18-24, generally spend more time on social media compared to older age groups.
Users from countries with higher internet penetration tend to have greater social media usage.

**2. Productivity Loss:**

There is a clear link between time spent on social media and productivity loss.
Increased social media usage is associated with greater productivity loss, particularly among younger individuals and those in professional or academic environments.
The negative impact on productivity is especially noticeable during work hours and late at night, suggesting that social media use during these times disrupts productivity.

**3. Platform Satisfaction:**

User satisfaction varies across different social media platforms, with some platforms receiving more favorable feedback than others.
Professional networking platforms often have higher satisfaction ratings among users aged 25-34.
Entertainment-focused platforms are more popular among users aged 18-24.
Platforms that offer localized content and language support tend to receive higher satisfaction ratings in non-English speaking countries.
Time of Day Impact:

Productivity loss is influenced by the time of day, with certain times showing more significant declines.
The greatest productivity losses occur late at night (10 PM - 2 AM) and in the mid-afternoon (2 PM - 4 PM).
These trends are consistent across different age groups and countries, indicating a widespread pattern of productivity dips linked to increased social media use during these periods.
